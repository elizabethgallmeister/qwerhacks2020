```python
import sys
assert sys.version_info >= (3, 5) # python>=3.5
import sklearn
assert sklearn.__version__ >= "0.20" # sklearn >= 0.20

import numpy as np #numerical package in python
import os
%matplotlib inline
import matplotlib.pyplot as plt #plotting package

# to make this notebook's output identical at every run
np.random.seed(42)

#matplotlib magic for inline figures
%matplotlib inline 
import matplotlib # plotting library
import matplotlib.pyplot as plt

# Where to save the figures
ROOT_DIR = "."

import pandas as pd

crime = pd.read_csv("Crime_Data_from_2010_to_Present.csv")
```

```python
crime.info()
crime.head()
```

```python
#looking for duplicate cases marked pts 1 and 2
crime.loc[crime["Date Rptd"]=="08/09/2010 12:00:00 AM"]
```

```python
#hist of victim age
crime["Vict Age"].hist(bins=20)
plt.show()
```

```python
#get rid of irrelevant categories
updatedCrime=crime.drop(columns=['DR_NO', 'Date Rptd', 'AREA NAME', 'Rpt Dist No', 'Part 1-2', 'Crm Cd Desc', 'Mocodes', 'Premis Desc', 'Weapon Desc', 'Status Desc', 'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street', 'Weapon Used Cd'])
updatedCrime.head()
```

```python
updatedCrime["Vict Sex"].value_counts()
```

```python
updatedCrime.dropna(subset = ["Vict Sex"], inplace=True)
updatedCrime.dropna(subset = ["Vict Descent"], inplace=True)
updatedCrime.dropna(subset = list(updatedCrime), inplace=True)

updatedCrime.info()
```

```python
#percentage of og data set remaining w/o null value cases
1913591/2110218
```

```python
updatedCrime.plot(kind="scatter", x="LON", y="LAT", alpha=0.1)
```

```python
#got rid of lat/long values of 0
upCrime = updatedCrime[updatedCrime.LAT != 0.0]
upCrime.info()
```

```python
#representation w/o lat/long vals
1912776/2110218
upCrime
```

```python
#attempted rape plot
spicyCrime = upCrime[upCrime["Crm Cd"] == 122]
spicyCrime.plot(kind="scatter", x = "LON", y = "LAT", alpha=0.3)
```

```python
#bounds of attempted rape scatterplot
lonMin = spicyCrime["LON"].min()
lonMax=spicyCrime["LON"].max()
latMin=spicyCrime["LAT"].min()
latMax=spicyCrime["LAT"].max()
print(lonMin, lonMax,latMin,latMax)
```

```python
#attempted rape on map
import matplotlib.image as mpimg

spicyCrime = upCrime[upCrime["Crm Cd"] == 122]
spicyCrime.plot(kind="scatter", x = "LON", y = "LAT", alpha=0.08)
la_image=mpimg.imread("./la.png")
plt.imshow(la_image, extent=[-118.6427 ,-118.1701 ,33.7105 ,34.3226], alpha=0.9,
           cmap=plt.get_cmap("jet"))
```

```python
#attempted rape by gender
import seaborn as sns
#sns.swarmplot(x="LON", y="LAT", data=spicyCrime)
sns.lmplot( x="LON", y="LAT", data=spicyCrime, fit_reg=False, hue='Vict Sex', legend=False, palette=dict(M="#9b59b6", F="#3498db"))
plt.legend(loc='lower left')
```

```python
#all crimes by descent
sns.lmplot( x="LON", y="LAT", data=upCrime, fit_reg=False, hue='Vict Descent', legend=False)
plt.legend(loc='lower left')
```
